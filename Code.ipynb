```python
# operate calculations on matrix 
import numpy as np 

# support for image processing such as : resize, homograpy,  click_event , draw circle on image 
import cv2

# compute euclidean distance 
from scipy.spatial import distance as dist 
```

```python
#resize the image for display 
def preprocessing(img):
    #width of new image 
    width = 1080 
    
    #height  of new image 
    height = 720
    
    #dim of the image
    dim = (width, height)
    resized= cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return resized
```

```python
#resize the image for display 
def afterprocessing(img):
    #width of new image 
    width = 800
    
    #height  of new image 
    height = 600
    
    #dim of the image
    dim = (width, height)
    resized= cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return resized
```

```python
#Source: https://www.pyimagesearch.com/2014/08/25/4-point-opencv-getperspective-transform-example/
#Goal:  
        #Apply homography on input image with 4 source points selected      
#Purpose:  
        # Computing  4 points of the output image base on selected points, pass them into 
        # cv2.getPerspectiveTransform() to compute 
        # transformation maxtrix and wrap it to display
    
    
def homography(image, pts):
    # obtain a consistent order of the points and unpack them individually
    rect = np.array(pts)
    (tl, tr, br, bl) = rect

    # compute the width of the new image, which will be the
    # maximum distance between bottom-right and bottom-left
    # x-coordiates or the top-right and top-left x-coordinates

    widthA = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))
    widthB = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
    maxWidth = max(int(widthA), int(widthB))

    # compute the height of the new image, which will be the
    # maximum distance between the top-right and bottom-right
    # y-coordinates or the top-left and bottom-left y-coordinates

    heightA = np.sqrt(((tr[0] - br[0]) ** 2) + ((tr[1] - br[1]) ** 2))
    heightB = np.sqrt(((tl[0] - bl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
    maxHeight = max(int(heightA), int(heightB))

    # now that we have the dimensions of the new image, construct
    # the set of destination points to obtain a "birds eye view",
    # (i.e. top-down view) of the image, again specifying points
    # in the top-left, top-right, bottom-right, and bottom-left
    # order
    dst = np.array([
        [0, 0],
        [maxWidth - 1, 0],
        [maxWidth - 1, maxHeight - 1],
        [0, maxHeight - 1]], dtype = "float32")
    
    # compute the perspective transform matrix and then apply it
    M = cv2.getPerspectiveTransform(rect, dst)
    warped = cv2.warpPerspective(image, M, (maxWidth, maxHeight))
    # return the warped image
    return warped
```

```python
#Source: https://www.geeksforgeeks.org/python-opencv-cv2-circle-method/
       # https://github.com/jialinding/ChessVision/blob/master/board.py 
#Goal:  Select 4 points on input image and store their coordinates      
#Purpose: Computing transformation matrix 
        
#initial array to store 4 point coordinates
src_pts = np.zeros( (4,2), np.float32)
#number of points slected
numPts=0

def click_event(event, x, y, flags, params): 
    global src_pts, numPts
    
    # end selecting when 4 points selected 
    if numPts > 3 :
        return
    # cv2.EVENT_LBUTTONDOWN is used for left_mouse_click
    if event == cv2.EVENT_LBUTTONDOWN: 
        # displaying the coordinateson the Shell 
        print(x,' ',y)
        
        # assign coordinates
        src_pts[numPts,0]=x
        src_pts[numPts,1]=y
        numPts = numPts + 1
        
        #Draw circle when clicking 
        # Center coordinates of circle
        center_coordinates = (x,y)
        
        # Radius of circle
        radius = 10
        
        # Red color in BGR
        color = (0, 0, 255)
        
        # Line thickness of -1 px
        thickness = -1
        
        cv2.circle(image, center_coordinates, radius, color,thickness)
        cv2.imshow("image",image)
```

```python
# driver function 
if __name__=="__main__": 
    
    #Step 1: Read and preprocesing 
    image=cv2.imread("Test 10 image\\top-view.jpg",1)
    if(image.shape[1] > 1080 and image.shape[0] > 720):
        image=preprocessing(image)
    clone_image= image.copy()
    
    #Step 2: Display and select manually 4 points of chessboard on source image for transformation matrix 
    cv2.imshow("image",image) 
    cv2.setMouseCallback("image",click_event) # call func click_event to select four conners of board 
    cv2.waitKey(0)
    
    #Step 3: # Apply homography on input image with 4 source points selected 
    out_image=homography(clone_image,src_pts)
    out_image=afterprocessing(out_image)
    #cv2.imwrite("Test 10 image\\top-view-topview.jpg", out_image)
    
    #Step 4: # Display output image 
    cv2.imshow("image", out_image)
    cv2.waitKey(0)
    
```
